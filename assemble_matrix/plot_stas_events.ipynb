```python
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
import numpy as np
```

```python
all_stas_isc = np.loadtxt("/Users/mary/Desktop/Rebuttal/assemblematrix/assemblematrix.stations.ISC_all", skiprows=2, dtype="object")
# all_stas_isc = all_stas[all_stas[:, 1] == 'ISC']
# all_stas_non_isc = all_stas[all_stas[:, 1] != 'ISC']
all_stas_pdiff = np.loadtxt("/Users/mary/Desktop/Rebuttal/assemblematrix/assemblematrix.stations.pdiff", skiprows=2, dtype="object")

all_eves_isc = np.loadtxt("/Users/mary/Desktop/Rebuttal/assemblematrix/assemblematrix.events.ISC_all", skiprows=2, dtype="object")
all_eves_pdiff = np.loadtxt("/Users/mary/Desktop/Rebuttal/assemblematrix/assemblematrix.events.pdiff", skiprows=2, dtype="object")

all_stas_rr = np.loadtxt("/Users/mary/Desktop/Rebuttal/assemblematrix/assemblematrix.stations.P_RR", skiprows=2, dtype="object")
all_eves_rr = np.loadtxt("/Users/mary/Desktop/Rebuttal/assemblematrix/assemblematrix.events.P_RR", skiprows=2, dtype="object")

```

```python
print(min(all_eves_isc[:,0]), max(all_eves_isc[:,0]), len(all_eves_isc))
print(min(all_eves_rr[:,0]), max(all_eves_rr[:,0]), len(all_eves_rr[:,0]))
print(min(all_eves_pdiff[:,0]), max(all_eves_pdiff[:,0]), len(all_eves_pdiff))
```

```python
# Sort so stations with highest number of measurements would be plotted on top
all_stas_isc = all_stas_isc[all_stas_isc[:, 2].astype(np.float).argsort()]
all_stas_rr = all_stas_rr[all_stas_rr[:, 2].astype(np.float).argsort()]
all_stas_pdiff = all_stas_pdiff[all_stas_pdiff[:, 2].astype(np.float).argsort()]

# all_stas_non_isc = all_stas_non_isc[all_stas_non_isc[:, 2].astype(np.float).argsort()]

all_eves_isc = all_eves_isc[all_eves_isc[:, 8].astype(np.float).argsort()]
all_eves_pdiff = all_eves_pdiff[all_eves_pdiff[:, 8].astype(np.float).argsort()]
all_eves_rr = all_eves_rr[all_eves_rr[:, 8].astype(np.float).argsort()]
```

```python
# ISC station
plt.figure(figsize=(30,15))

m = Basemap(projection='robin',lon_0=0, resolution='c')
#m.drawcoastlines()
m.drawparallels(np.arange(-90.,120.,30.))
m.drawmeridians(np.arange(0.,360.,60.))
#m.drawmapboundary(fill_color=None)
m.fillcontinents(color='lightgray')


# Define cmap
cmap = plt.get_cmap('jet', 10)

x, y = m(all_stas_isc[:, 7].astype(np.float), 
         all_stas_isc[:, 6].astype(np.float))

im = plt.scatter(x, y, 20, 
                 c=all_stas_isc[:, 2].astype(np.float), 
                 marker='v', 
                 cmap=cmap, 
                 vmin=0, 
                 vmax=200, zorder=2)

# im = plt.scatter(x, y, 16, 
#                  c='red', 
#                  marker='v', alpha=0.1)

cb = plt.colorbar(im, fraction=0.02, pad=0.01, orientation="horizontal")
cb.ax.tick_params(labelsize=24)
plt.savefig('/Users/mary/Desktop/Rebuttal/assemblematrix/ISC_stations.png', dpi=300)
```
